# Extarction of Top 10 Countries With Respect To Total Population:

```python
# Step 1: Filter for population indicator
population_df = df[df['Indicator Name'] == 'Population, total']

# Step 2: Drop missing values and convert 2022 to numeric
population_df = population_df[[ 'Country Code', '2022']].dropna()
population_df['2022'] = pd.to_numeric(population_df['2022'], errors='coerce')

# Step 3: Get top 10 by population
top_10 = population_df.sort_values(by='2022', ascending=False).head(10)

# Step 4: Remove the 2022 column if no longer needed
top_10 = top_10.drop(columns=['2022'])

# Step 5: Print the result
print("Top 10 Countries by Population:")
print(top_10)
```

# BAR PLOT:

# Top Ten Countries of Total Population in Year 2022 and 2016:

```python
# Filter for "Population, total"
pop_df = df[df['Indicator Name'] == 'Population, total']

# Remove aggregates: keep only rows where 'Country Code' is 3 letters (most countries)
pop_countries = pop_df[pop_df['Country Code'].str.len() == 3]

# Convert population columns to numeric
pop_countries['2016'] = pd.to_numeric(pop_countries['2016'], errors='coerce')
pop_countries['2022'] = pd.to_numeric(pop_countries['2022'], errors='coerce')

# Get top 10 countries for 2016
top10_2016 = pop_countries.nlargest(10, '2016')[['Country Code', '2016']]

# Get top 10 countries for 2022
top10_2022 = pop_countries.nlargest(10, '2022')[['Country Code', '2022']]

# Plot side by side
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.barplot(x='2016', y='Country Code', data=top10_2016, palette='coolwarm')
plt.title('Top Ten Countries by Total Population (2016)', fontsize=12)
plt.xlabel('Total Population', fontsize=10)
plt.ylabel('Country', fontsize=10)

plt.subplot(1, 2, 2)
sns.barplot(x='2022', y='Country Code', data=top10_2022, palette='coolwarm')
plt.title('Top Ten Countries by Total Population (2022)', fontsize=12)
plt.xlabel('Total Population', fontsize=10)
plt.ylabel('Country', fontsize=10)

plt.tight_layout()
plt.show()
```

# Extraction of bottom 10 countries with respect to total popultaion:

```python
# Filter for "Population, total"
pop_df = df[df['Indicator Name'] == 'Population, total']

# Remove aggregates: keep only rows where 'Country Code' is 3 letters (most countries)
pop_countries = pop_df[pop_df['Country Code'].str.len() == 3]

# Convert the 2022 column to numeric
pop_countries['2022'] = pd.to_numeric(pop_countries['2022'], errors='coerce')

# Sort data based on the total population for 2022 (ascending order)
total_population_sorted1 = pop_countries.sort_values(by="2022", ascending=True)

# Get the bottom ten countries with the lowest total population for 2022
total_bottom_ten_countries = total_population_sorted1.head(10)

# Check columns
print("Columns in DataFrame:", total_bottom_ten_countries.columns.tolist())

# Print the desired columns (make sure they exist)
print("Bottom ten countries of total population (2022)\n")
print(total_bottom_ten_countries[['Country Name', 'Country Code', '2022']])
```

# Bottom ten countries of total population in year 2022 and 2016:

```python
# Filter for "Population, total"
pop_df = df[df['Indicator Name'] == 'Population, total']

# Keep only actual countries (country codes are 3 letters)
pop_countries = pop_df[pop_df['Country Code'].str.len() == 3]

# Convert population columns to numeric
pop_countries['2016'] = pd.to_numeric(pop_countries['2016'], errors='coerce')
pop_countries['2022'] = pd.to_numeric(pop_countries['2022'], errors='coerce')

# Get the bottom ten countries by population for 2016 and 2022
bottom10_2016 = pop_countries.nsmallest(10, '2016')[['Country Code', '2016']]
bottom10_2022 = pop_countries.nsmallest(10, '2022')[['Country Code', '2022']]

# Plot both in the same figure
plt.figure(figsize=(15, 6))

# 2016 subplot
plt.subplot(1, 2, 1)
sns.barplot(x='2016', y='Country Code', data=bottom10_2016, palette='coolwarm')
plt.title('Bottom Ten Countries by Total Population (2016)', fontsize=12)
plt.xlabel('Total Population', fontsize=10)
plt.ylabel('Country', fontsize=10)

# 2022 subplot
plt.subplot(1, 2, 2)
sns.barplot(x='2022', y='Country Code', data=bottom10_2022, palette='coolwarm')
plt.title('Bottom Ten Countries by Total Population (2022)', fontsize=12)
plt.xlabel('Total Population', fontsize=10)
plt.ylabel('Country', fontsize=10)

plt.tight_layout()
plt.show()
```

# Extarction of top 10 countries with highest male population:

```python
# Remove aggregates: keep only rows where 'Country Code' is 3 letters (most countries)
pop_countries = df[(df['Indicator Code'] == 'SP.POP.TOTL') & (df['Country Code'].str.len() == 3)]

# Convert the 2022 column to numeric
pop_countries['2022'] = pd.to_numeric(pop_countries['2022'], errors='coerce')

# Sort data based on the total population for 2022 (descending order)
pop_sorted = pop_countries.sort_values(by="2022", ascending=False)

# Get the top 10 countries with the highest total population for 2022
top_ten_countries = pop_sorted.head(10)

print("Top ten countries of total population (2022):")
print(top_ten_countries[['Country Name', 'Country Code', '2022']])
```

```python
# Choose the country
chosen_country = 'United States'

# Filter for total population indicator and the chosen country
country_data = df[
    (df['Country Name'] == chosen_country) &
    (df['Indicator Code'] == 'SP.POP.TOTL')
]

# Select only the year columns (from 1960 to 2024)
year_cols = [col for col in country_data.columns if col.isdigit()]
years = list(map(int, year_cols))
population = country_data[year_cols].values.flatten()

# Convert population to numeric (in case of missing/empty values)
population = pd.to_numeric(population, errors='coerce')

# Plot
plt.figure(figsize=(12, 5))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.tight_layout()
plt.show()
```
